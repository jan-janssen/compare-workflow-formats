```python
from jobflow import job, Flow
from jobflow.managers.local import run_locally
```

```python
@job
def add_xy(x, y):
    return x + y
```

```python
@job
def add_xyz(x, y, z):
    return x + y + z
```

```python
x = 1
y = 2
z = add_xy(x=x, y=y)
w = add_xyz(x=x, y=y, z=z.output)
```

```python
flow = Flow([z, w])
```

```python
flow.as_dict()
```

```python
result = run_locally(flow)
result
```
